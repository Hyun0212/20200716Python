```python
text = open(r'C:\Users\user\Desktop\302\thelittleprince.txt').read()
```

```python
img = r'C:\Users\user\Desktop\302\1.png'
```

```python
Image.open(img)
```

```python
lp_mask=np.array(Image.open(img))
lp_mask
```

```python
lp_wc=WordCloud(
    background_color='white',
    mask=lp_mask,
    max_font_size=50,
    contour_width=5,
    contour_color='green').generate(text)
lp_wc.words_
lp_wc.to_image
lp_wc.to_file(r'C:\Users\user\Desktop\302/1.png')    
```
